```python
import pandas as pd
import numpy as np

# Assuming your dataset is in a DataFrame named 'df'

# Create a copy of the DataFrame
df_cleaned = df.copy()

# Iterate over each column
for column_name in df.columns:
    # Calculate the quartiles and interquartile range (IQR)
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Determine the upper and lower bounds for capping
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    # Cap outliers using np.where()
    df_cleaned[column_name] = np.where(
        df_cleaned[column_name] > upper_bound,
        upper_bound,
        np.where(
            df_cleaned[column_name] < lower_bound,
            lower_bound,
            df_cleaned[column_name]
        )
    )

# Reset the index of the cleaned DataFrame
df_cleaned = df_cleaned.reset_index(drop=True)
```
